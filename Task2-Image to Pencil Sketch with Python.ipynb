## Task 2: Image to Pencil Sketch with Python

### Importing Libraries

```python
import cv2
from PIL import Image
from IPython.display import display
```

### Loading Image

Resizing and converting Image into RGB format

```python
img = cv2.imread('TaylorSwift.jpg',1)
img = cv2.resize(img, (400,250))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

### Converting Image into Gray Image

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

### Inverting the Image

```python
inverted_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(inverted_image))
```

### Blurring the Image using Gaussian Blur

```python
blurred = cv2.GaussianBlur(inverted_image, (5, 5), 0, 0)
display(Image.fromarray(blurred))
```

### Inverting the Image to get the Final Sketch

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=250.0)
display(Image.fromarray(pencil_sketch))
```

### Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(img))
display(Image.fromarray(pencil_sketch))
```
